```python
#Import the pandas library to work with data
import pandas as pd
```

```python
#Read the data from  the csv file 'Iris.csv' and store it in 'data' dataframe
data=pd.read_csv('Iris.csv')
```

```python
#Display entire 'data' dataframe
data
```

```python
#Generating a boolean series indicating whether each row in the dataframe is duplicate or not
data.duplicated()
```

```python
#Calculate the total no. of duplicate rows in the datafame
data.duplicated().sum()
```

```python
#Calculate the total no. of duplicate rows in the  'SepalLengthCm' column
data['SepalLengthCm'].duplicated().sum()
```

```python
#Retrieve the first 20 rows where the 'SepalLengthCm' column contain duplicate values
data[data['SepalLengthCm'].duplicated()].head(20)
```
