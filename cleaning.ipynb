```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
data
```

```python
data.isnull().sum()#standard missing values
```

```python
data.replace(0,np.NaN,inplace=True)
data.head(20)
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
mean=round(data["Survived"].mean(),2)
mean
```

```python
data['Survived'].fillna(mean,inplace=True)
data['Survived'].head(20)
```

```python
data['Age'].interpolate(inplace=True)
data['Age'].head(20)
```

```python
data['Age'].bfill(inplace=True)
data['Age'].head(20)
```

```python
data['SibSp'].ffill(inplace=True)
data['SibSp'].head(20)
```

```python
data['Parch'].fillna(3.0,inplace=True)
data['Parch'].head(20)
```

```python
mode=data['Cabin'].mode()
mode
```

```python
mode=data['Cabin'].mode()[1]
mode
```

```python
data['Cabin'].fillna(mode,inplace=True)
data['Cabin'].head(20)
```

```python
data['Embarked'].bfill(inplace=True)
data['Embarked'].head(20)
```
