```python
#importing the required libraries
import numpy as np
import pandas as pd
import statistics as stats
```

```python
#Reading the dataset from the'train.csv' file and storing it in a 'data' variable
data=pd.read_csv('train.csv')
data
```

```python
#calculating the mean of the entire dataset and printing  the result
mean=np.mean(data)
print('MEAN:\n', mean)
```

```python
#calculating the mean of the 'Age' column in the dataset
data.loc[:,'Age'].mean()
```

```python
#Calculating the median of the'PassengerId' and 'Fare' columns in the dataset
data[['PassengerId','Fare']].median()
```

```python
#Finding the mode of the entire dataset
data.mode()
```

```python
#Finding the mode of the 'Ticket' column in the dataset
data['Ticket'].mode()
```

```python
#Calculating the standard deviation of the entire dataset
std=np.std(data)
print(std)
```

```python
#Counting the number of non-null values in the 'Fare' column of the dataset
data.count()
```

```python
#Finding the maximum value in the 'Fare' column of the dataset
data['Fare'].max()
```
