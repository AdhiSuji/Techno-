```python
#Importing the required libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the data from 'Iris.csv' file and stored it in a variable 'data'
data=pd.read_csv('Iris.csv',delimiter=',')
```

```python
#Printing the columns 'SepalLengthCm','SepalWidthCm','PetalLengthCm' and 'PetalWidthCm' from the dataFrame 'data'
print(data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
```

```python
#Plotting a histogram of the 'SeapalLengthCm' column
plt.hist(data['SepalLengthCm'],color='b')
plt.title('SEPAL LENGTH IN CM')
plt.xlabel('Cm')
plt.ylabel('Count')
plt.show()
```

```python
#Plotting a histogram of the 'SepalWidthCm' column
plt.hist(data['SepalWidthCm'],color='c')
plt.title('SEPAL WIDTH IN CM')
plt.xlabel('cm')
plt.ylabel('count')
plt.show()
```

```python
#Plotting a histogram of the 'PetalLengthCm' column
plt.hist(data['PetalLengthCm'],color='g')
plt.title('PETAL LENGTH IN CM')
plt.xlabel('cm')
plt.ylabel('count')
plt.show()
```

```python
#Plotting a histogram of the 'PetalWidthCm' column
plt.hist(data['PetalWidthCm'],color='r')
plt.title('PETAL WIDTH IN CM')
plt.xlabel('Cm')
plt.ylabel('count')
plt.show()
```

```python
#Plotting a barchart of the 'Species' column
sns.countplot(data['Species'])
plt.title("Iris Species Count")
plt.show()
```
